```python
# Removal of Duplicate values
rs,cs = df.shape
df.drop_duplicates(inplace=True)
if df.shape==(rs,cs):
    print('\n The dataset doesn\'t have any duplicates')
else:
    print(f'\n Number of duplicates dropped/fixed ---> {rs-df.shape[0]}')
```

```python
df["Area Type"] = df["Area Type"].map({"Super Area": 1, 
                                           "Carpet Area": 2, 
                                           "Built Area": 3})
df["City"] = df["City"].map({"Mumbai": 4000, "Chennai": 6000, 
                                 "Bangalore": 5600, "Hyderabad": 5000, 
                                 "Delhi": 1100, "Kolkata": 7000})
df["Furnishing Status"] = df["Furnishing Status"].map({"Unfurnished": 0, 
                                                           "Semi-Furnished": 1, 
                                                           "Furnished": 2})
df["Tenant Preferred"] = df["Tenant Preferred"].map({"Bachelors/Family": 2, 
                                                         "Bachelors": 1, 
                                                         "Family": 3})
print(df.head())
```

```python
# CONVERTING ALL OUR CATEGORICAL DATA COLUMNS TO NUMERIC FORM

from sklearn.preprocessing import LabelEncoder
categ = ["Posted On","Floor","Area Type","Area Locality","City","Furnishing Status","Tenant Preferred","Point of Contact"]

# Encode Categorical Columns
le = LabelEncoder()
df[categ] = df[categ].apply(le.fit_transform)
```

```python
# After converting the Data looks like
df.head(15)
```

```python
# Corelation Between Columns
corr = df.corr()
plt.figure(figsize=(12,7))
sns.heatmap(corr,cmap='coolwarm',annot=True)
```

```python
# Splitting the Dataset
x = df.drop(['Rent'],axis=1)
y = df['Rent']
```

```python
# Data after Dropping Rent
x
```

```python
# Rent
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
# LENGTH OF x_train AND x_test

len(x_train),len(x_test)
```

```python
#Linear Regression :

# IMPORTING THE MODULE

from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
# FITTING THE DATA INTO THE MODEL

model.fit(x_train,y_train)
```

```python
# PREDICTING THE OUTCOMES

y_predict = model.predict(x_test)
```

```python
y_predict
```

```python
#Evaluation :

from sklearn.metrics import r2_score,mean_absolute_error
score = r2_score(y_test,y_predict)
mae = mean_absolute_error(y_test,y_predict)
```
